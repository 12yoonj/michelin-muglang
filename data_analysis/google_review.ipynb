```python
cat = """user
churches
resorts
beaches
parks
theatres
museums
malls
zoo
restaurants
pubs/bars
local services
burger/pizza shops
hotels/other lodgings
juice bars
art galleries
dance clubs
swimming pools
gyms
bakeries
beauty & spas
cafes
view points
monuments
gardens"""
columns_list = cat.split("\n")
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data/google_review_ratings.csv', names=columns_list, header=0,index_col=False)
data = data.set_index('user')
```

```python
review = data
print(review.shape)
review.head()
```

```python
review.info()
```

```python
# 결측치 확인
review.isnull().sum()
```

```python
# 사용할 열만 추출
review = review.loc[:, ['beaches', 'parks', 'theatres', 'malls', 'zoo', 'restaurants', 'art galleries', 'swimming pools', 'beauty & spas', 'view points', 'monuments']]
review.head()
```

```python
# 카테고리별 평점 평균
mean_review = review.mean().sort_values(ascending=False)
mean_review
```

```python
mean_review = mean_review.reset_index()
mean_review
```

```python
mean_review.rename(columns={'index':'place', 0: 'rating'}, inplace=True)
mean_review
```

```python
plt.figure(figsize=(15, 8))
sns.barplot(data=mean_review, x='rating' , y='place')
plt.show()
```
