```python
data = michelin.copy()
```

```python
data['imgURL'] = np.nan
data.head()
```

```python
lenData = data.shape[0]
lenData
```

```python
from urllib.request import HTTPError
```

```python
for i in range(lenData):
    try:
        html = urlopen(data['Url'][i])
    except HTTPError as e:
        print(f'{i} page not found')
        data['imgURL'][i] = []
    else:
        soup = BeautifulSoup(html, "html.parser") 

        lstLen = len(list(soup.find_all("div", {"data-target":"#js-gallery-masthead"})))
        lst = [soup.find_all("div", {"data-target":"#js-gallery-masthead"})[j].get('data-bg') for j in range(lstLen)]
        data['imgURL'][i] = lst
```

```python
data.to_csv('data_end.csv',index=False)
```
